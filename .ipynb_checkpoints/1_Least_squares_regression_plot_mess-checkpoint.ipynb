```python
import matplotlib.pyplot as plt
import random
import numpy as np
import time
from copy import deepcopy
from sklearn.datasets import make_regression as Make
from sklearn.datasets import load_diabetes as Diabetes
from sklearn.datasets import load_boston as Boston
from sklearn.datasets import fetch_california_housing as California
```

```python
class leastSquaresGD:
    def __init__(self, error, maxTime):
        # original copies to reference back to
        self.x_original = None  
        self.a_original = None
        self.b_original = None
        # the copies
        self.x = None      # weights
        self.a = None      # xTrain
        self.b = None      # coeff
        self.initialTime = None # starting time
        self.error = error   # epsilon
        self.grad = None    # current gradient
        self.norm = None    # current norm of gradient
        self.lam = None     # current step size
        self.gr = None      # list of values of norms of gradient
        self.lam_list = None # # list of values of step sizes
        self.timeAxis = None # timeAxis[x] is the time at gr[x]
        self.theta = None   # theta value used in Adpative
        # Plots for different methods
        # syntax is 1. Iter v Norm, 2. Iter v Lambda, 3. Time v Norm, 4. Time v Lambda
        self.vanPlt = [None, None, None, None] 
        self.AdapPlt = [None, None, None, None]
        self.LSPlt = [None, None, None, None]
        self.BazPlt = [None, None, None, None]
        self.Max = 0         # maximum value for x axis in plt
        self.maxRunTime = maxTime
        self.time = None
        self.lam_list = None
    
    def get_ds(self, ds, Import = True):
        # can use 'make_regression', 'load_boston', 'fetch_california_housing', 'load_diabetes'
        if Import == True:
            dataset = ds(return_X_y=True)
        else:
            dataset = ds()
        self.a_original = dataset[0]
        self.b_original = dataset[1]
        self.x_original = np.random.rand(len(self.a_original[0]))
    
    def generate(self, function, dim, vec_len, scale):
        # A function to generate the data for the regression
        for i in range(dim):
            globals()['X'+str(i)] = [0]*vec_len
        for j in range(vec_len):
            globals()['X'+str(0)][j] = (random.random()+1e-12)*scale
            y = function(globals()['X'+str(0)][j])
            noiz = 3
            for k in range(dim):
                globals()['X'+str(k)][j] = y + np.random.normal(0,noiz)
                noiz += 3
        A = np.array([globals()['X'+str(i)] for i in range(dim)]).T
        self.a_original = A
        self.x_original = np.random.rand(len(A[0]))*scale
        self.b_original = A @ self.x_original + np.random.normal(0,3,(A @ self.x_original).shape)
        
    def f(self, X):
        return 0.5 * np.linalg.norm(self.a @ X - self.b)**2
    def g(self, X):
        return self.a.T @ (self.a @ X - self.b)
    def Step(self):
        self.x = self.x - self.lam*self.grad
    
    def normal(self, X): 
        return np.linalg.norm(X)
    
    def Vanilla(self, timeAx = False):
        self.initialTime = time.time()
        self.x = deepcopy(self.x_original)
        self.a = deepcopy(self.a_original)
        self.b = deepcopy(self.b_original)
        self.gr = []
        self.timeAxis = []
        self.lam_list = []
        L = np.max( np.linalg.svd( self.a @ self.a.T ) [1] )
        self.lam = 1/L
        self.grad = self.g(self.x)
        self.norm = self.normal(self.grad)
        while self.norm > self.error:
            self.Step()
            self.grad = self.g(self.x)
            self.norm = self.normal(self.grad)
            self.gr.append(self.norm)
            self.timeAxis.append(time.time() - self.initialTime)
            self.lam_list.append(self.lam)
            if time.time() - self.initialTime > self.maxRunTime: # check within time limit
                break
        self.time = round(time.time() - self.initialTime, 2) # stop the clock
        print(' \n Vanilla Gradient Descent:')
        print('_____________________________________________')
        if len(self.gr) > self.Max:
            self.Max = deepcopy(len(self.gr))
        if timeAx == True:
            self.SetTimePlots(0, len(self.gr) , 'g', self.vanPlt, 'Vanilla {} sec'.format(self.time))
        else:
            self.SetIterPlots(0, min(len(self.gr),600) , 'g', self.vanPlt, 'Vanilla {} sec'.format(self.time))
        self.PrintResults()
    
    def Adaptive(self):
        self.initialTime = time.time()
        self.x = deepcopy(self.x_original)
        self.a = deepcopy(self.a_original)
        self.b = deepcopy(self.b_original)
        self.gr = []
        self.timeAxis = []
        self.lam_list = []
        self.lam = random.random()+1e-12
        self.theta = float(random.getrandbits(128))
        oldX = deepcopy(self.x)
        self.grad = self.g(self.x)
        self.gr.append(self.norm)
        self.timeAxis.append(time.time() - self.initialTime)
        self.lam_list.append(self.lam)
        self.Step()
        self.norm = self.normal(self.grad)
        self.gr.append(self.norm)
        self.timeAxis.append(time.time() - self.initialTime)
        self.lam_list.append(self.lam)
        while self.norm > self.error:
            oldLam = deepcopy(self.lam)
            min1 = np.sqrt(1 + self.theta)*self.lam
            min2 = ( np.linalg.norm( self.x - oldX ) ) \
                / ( 2 * np.linalg.norm(self.g(self.x) - self.g(oldX) ) )
            self.lam = np.min([min1,min2])
            self.step_list.append(self.lam)
            oldX = deepcopy(self.x)
            self.Step() # update x
            self.theta = self.lam/oldLam
            self.grad = self.g(self.x)
            self.norm = self.normal(self.grad)
            self.gr.append(self.norm)
            self.timeAxis.append(time.time() - self.initialTime)
            self.lam_list.append(self.lam)
            if time.time() - self.initialTime > self.maxRunTime: # check within time limit
                break
        self.time = round(time.time() - self.initialTime, 2) # stop the clock
        print('\n Adaptive Gradient Descent:')
        print('_____________________________________________')
        if len(self.gr) > self.Max:
            self.Max = deepcopy(len(self.gr))
        self.SetIterPlots(0, min(len(self.gr),600) , 'y', self.AdapPlt, 'Adaptive {} sec'.format(self.time))
        self.PrintResults()
    
    def BackTrack(self):
        rho = random.random()
        lam = random.randrange(1e6)
        while lam > ( self.normal( lam * self.grad ) / \
                     ( self.normal( self.g( self.x - lam * self.grad ) - self.grad ) )  ):
            lam = rho * lam
        self.lam = lam

    def LineSearch(self):
        self.initialTime = time.time()
        self.x = deepcopy(self.x_original)
        self.a = deepcopy(self.a_original)
        self.b = deepcopy(self.b_original)
        self.gr = []
        self.timeAxis = []
        self.lam_list = []
        self.grad = self.g(self.x)
        self.norm = self.normal(self.grad)
        self.BackTrack()
        self.gr.append(self.norm)
        self.timeAxis.append(time.time() - self.initialTime)
        self.lam_list.append(self.lam)
        while self.norm > self.error:
            self.BackTrack()
            self.Step()
            self.grad =  self.g(self.x)
            self.norm = self.normal(self.grad)
            self.gr.append(self.norm)
            self.timeAxis.append(time.time() - self.initialTime)
            self.lam_list.append(self.lam)
            if time.time() - self.initialTime > self.maxRunTime: # check within time limit
                break
        self.time = round(time.time() - self.initialTime, 2) # stop the clock
        print('\n Line Search Gradient Descent:')
        print('_____________________________________________')
        if len(self.gr) > self.Max:
            self.Max = deepcopy(len(self.gr))
        self.SetIterPlots(0, min(len(self.gr),600) , 'b', self.LSPlt, 'LineSearch {} sec'.format(self.time))
        self.PrintResults()
    
    def Barzilai(self):
        self.initialTime = time.time()
        self.x = deepcopy(self.x_original)
        self.a = deepcopy(self.a_original)
        self.b = deepcopy(self.b_original)
        self.gr = []
        self.timeAxis = []
        self.lam_list = []
        self.grad = self.g(self.x)
        self.lam = random.random()+1e-12
        oldX = deepcopy(self.x)
        self.Step()
        self.norm = self.normal(self.grad)
        while self.norm > self.error:
            oldLam = deepcopy(self.lam)
            self.lam = np.dot( self.x - oldX, self.g(self.x) - self.g(oldX) ) \
                        / np.linalg.norm(self.g(self.x) - self.g(oldX))**2
            oldX = deepcopy(self.x)
            self.Step()
            self.grad = self.g(self.x)
            self.norm =  self.normal(self.grad)
            self.gr.append(self.norm)
            self.timeAxis.append(time.time() - self.initialTime)
            self.lam_list.append(self.lam)
            if time.time() - self.initialTime > self.maxRunTime: # check within time limit
                break
        self.time = round(time.time() - self.initialTime, 2) # stop the clock
        print('\n Barzilai-Borwein Gradient Descent:')
        print('_____________________________________________')
        if len(self.gr) > self.Max:
            self.Max = deepcopy(len(self.gr))
            
        self.BazPlt[0] = range(len(self.gr))
        self.BazPlt[1] = self.gr
        self.BazPlt[2] = self.lam_list
        self.BazPlt[3] = self.timeAxis
        
        self.PrintResults()
    
#     def SetTimePlots(self, Min, Max, colour, name, label):
#         name = plt.plot( self.timeAxis[Min:Max] , self.gr[Min:Max] , c = colour , label = label)
    
#     def SetIterPlots(self, Min, Max, colour, name, label): #change this back
#         name = plt.plot( range( Min, Max ) , self.gr[Min:Max] , c = colour , label = label)

    def PrintPlots(self):
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)
        plt.yscale('log')

        ax1.plot(self.BazPlt[0], self.BazPlt[1], c = 'r', label = 'Barzilai')
        ax1.xlabel = 'Iterations'
        ax1.ylabel = 'Norm of Gradient'
        ax1.legend(loc='best')

        ax2.plot(self.BazPlt[0], self.BazPlt[2], c = 'r', label = 'Barzilai')
        ax2.xlabel = 'Iterations'
        ax2.ylabel = 'Step Size'
        ax2.legend(loc='best')

        ax3.plot(self.BazPlt[3], self.BazPlt[1], c = 'r', label = 'Barzilai')
        ax3.xlabel = 'Time (seconds)'
        ax3.ylabel = 'Norm of Gradient'
        ax3.legend(loc='best')

        ax4.plot(self.BazPlt[3], self.BazPlt[2], c = 'r', label = 'Barzilai')
        ax4.xlabel = 'Time (seconds)'
        ax4.ylabel = 'Norm of Gradient'
        ax4.legend(loc='best')
        plt.show()

#     def PrintPlots(self):
#         fig, ax = plt.subplots
#         ax.yscale('log')
        
#         plt.subplot(2,2,1)
#         plt.plot(self.BazPlt[0], self.BazPlt[1], c = 'r', label = 'Barzilai')
#         plt.xlabel = 'Iterations'
#         plt.ylabel = 'Norm of Gradient'
#         plt.legend(loc='best')
        
#         plt.subplot(2,2,2)
#         plt.plot(self.BazPlt[0], self.BazPlt[2], c = 'r', label = 'Barzilai')
#         plt.xlabel = 'Iterations'
#         plt.ylabel = 'Step Size'
#         plt.legend(loc='best')
        
#         plt.subplot(2,2,3)
#         plt.plot(self.BazPlt[3], self.BazPlt[1], c = 'r', label = 'Barzilai')
#         plt.xlabel = 'Time (seconds)'
#         plt.ylabel = 'Norm of Gradient'
#         plt.legend(loc='best')
        
#         plt.subplot(2,2,4)
#         plt.plot(self.BazPlt[3], self.BazPlt[2], c = 'r', label = 'Barzilai')
#         plt.xlabel = 'Time (seconds)'
#         plt.ylabel = 'Norm of Gradient'
#         plt.legend(loc='best')
        
# #         for (i,j,k) in zip(range(1,5), \
# #                            ['Iterations', 'Iterations', 'Time', 'Time'], \
# #                           ['Norm of Gradient', 'Step Size', 'Norm of Gradient', 'Step Size'] ):
# #             plt.subplot(2,2,i)
# #             self.vanPlt[i-1]
# #             self.AdapPlt[i-1]
# #             self.LSPlt[i-1]
# #             self.BazPlt[i-1]
# #             plt.xlabel = j
# #             plt.ylabel = k
# #             plt.legend(loc='best')
# #         plt.title(label = 'Norm of Gradient in domain [{}, {}]'.format(0,self.Max) )
#         plt.show()
    def PrintResults(self):
        print('At final iteration k =', len(self.gr), ':')
        print('x vector is:', self.x)
        print('Gradient Norm is:', self.norm)
        print('Time taken:', self.time)
        print('Final Lambda value: ', self.lam)
        
def linFunc(x):
    return 2*x + 5
```
