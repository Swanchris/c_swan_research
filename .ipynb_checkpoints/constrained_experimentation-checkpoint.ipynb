# Constrained Least Squares

```python
import matplotlib.pyplot as plt
import random
import numpy as np
import pandas as pd
import time
from copy import deepcopy
from sklearn.datasets import make_regression as Make_reg
from sklearn.model_selection import train_test_split
def ren(y): return range(len(y))
rng = np.random.default_rng()
```

```python
DS = Make_reg(n_samples=1000 ,n_features=10)
A1, A2, b1, b2 = train_test_split(DS[0],DS[1], test_size=0.1)
b1 = np.array([i+rng.uniform(-1,1)*10 for i in b1])
A2inv = np.linalg.pinv(A2)
```

Assume we have a dataset for which we want to do regression on. We know this dataset is divided into reliable and unreliable data, such that it takes the form 
$$\begin{align}
    A_1 x &\approx b_1\\
    A_2 x &= b_2
  \end{align}
$$
Consequently we have the program:
$$\begin{align}
    \min_x \quad &\frac{1}{2} \| A_1 x - b_1\|^2 \\
    \text{s.t.} \quad &A_2 x = b_2
  \end{align}
$$

We capture the constraint by projecting the gradient descent function onto the set $C$ which is defined as $$C = \{x \in \mathbb{R}^n : Ax = b\}.$$ The formula for the projection onto $C$ using the Moore-Penrose pseudo-inverse is $$P_C(z) = z - (A^{\intercal} A)^{-1} A^{\intercal}(Az - b).$$

Using the convention $$A^+ = (A^{\intercal} A)^{-1} A^{\intercal}$$ the above reduces to $$ P_C(z) = z - A^+(Az - b).$$


We thus have the following procedure:
$$\begin{align}
f(x) &= \frac{1}{2} \| A_1 x - b_1\|^2 \\
\nabla f(x) &= A_1^{\intercal} (A_1 x - b_1)\\
y^k &= x^k - \lambda \nabla f(x^k)\\
&= x^k - \lambda \nabla \big(A_1^{\intercal} (A_1 x^k - b_1)\big)\\
x^{k+1} &= P_C(y^k)\\
&= y^k - A_2^+ (A_2 y^k - b_2)\\
&= x^k - \lambda \nabla f(x^k) - A_2^+ \big(A_2(x^k - \lambda \nabla f(x^k)) - b_2\big)\\
&= x^k - \lambda \nabla \big(A_1^{\intercal} (A_1 x^k - b_1)\big) - (A_2^{\intercal} A_2)^{-1} A_2^{\intercal}\Bigg(A_2\bigg(x^k - \lambda \nabla \big(A_1^{\intercal} (A_1 x^k - b_1)\big)\bigg) - b_2\Bigg)
\end{align}$$

```python
def f(x,A,b):
    return 0.5*(np.linalg.norm( A @ x - b )**2)
def f_grad(x,A,b):
    return A.T @ ( A @ x - b )
def y(x,A,b,lam):
    return x - lam*f_grad(x,A,b)
```

```python
x = rng.random(len(A1[0]))
lam = 1/np.max(np.linalg.svd(A1@A1.T)[1])
y1 = y(x,A1,b1,lam)
```

# Old rubbish problem

```python
DS = Make_reg(n_samples=100 ,n_features=5)
# print(DS[0].T.shape)
# print(DS[1].shape)
A = DS[0].T
b = DS[1]
rng = np.random.default_rng()
x = rng.random(len(A))
def gamma(xx, AA, bb): return np.max((AA.T @ xx - bb)**2)
def f(xx, AA, bb): return 0.5*np.linalg.norm(AA.T @ xx - bb)**2 # for whole matrix A
def g(xx, AA, bb, gam, i): # For individual g_i functions
    g1 = np.linalg.norm(AA.T[i]@ xx - bb[i])**2 - gam
    g2 = np.max([0,g1])
    return g2
def h(xx,AA,bb, ep): return np.sum((AA.T @ xx - 0.5*bb + ep)**2) # for whole matrix A
def alpha(k): return k**2
def epsy(bb): return rng.random(len(bb))
def P(xx,AA,bb,gam,eps,k):
    f_part = f(xx,AA,bb)
    g_part = np.sum([g(xx,AA,bb,gam,i)**2 for i in ren(bb)])
    h_part = h(xx,AA,bb,eps)**2
    return f_part + 0.5*alpha(k)*( g_part + h_part )
def P_grad(xx, AA, bb, gam, eps, k):
    f_part = AA@(AA.T@xx -bb)
    g_part = np.sum([(AA.T[i]@xx - bb[i])*g(xx,AA,bb,gam,i)*AA.T[i] for i in ren(bb)], axis=0)
    h_part = h(xx,AA,bb,eps)*( AA@(AA.T@xx - 0.5*bb + eps) )
    return f_part + 2*alpha(k)*(g_part + h_part)
def GD_constrain(xx,AA,bb,gam,eps,k, LAMBDA):
    return xx - LAMBDA*P_grad(xx,AA,bb,gam,eps,k)
```
